```python
df1 = pd.DataFrame(columns = ['Area of residence', 'Area of workplace', 'On foot'])
df1['Area of residence'] = df['Area of residence']
df1['Area of workplace'] = df['Area of workplace']
df1['On foot'] = df['On foot']

df2 = df
```

```python
# Read the Boundaries of City of Toronto Neighbourhoods.
import numpy as np
import pandas as pd
import geopandas as gpd
import datetime as dt
import time
import matplotlib.pyplot as plt

# This process may take 3 seconds,
t1 = time.time()
url_geo = "Middle_Layer_Super_Output_Areas_(December_2011)_Boundaries_Generalised_Clipped_(BGC)_EW_V3.geojson"
print('Loading data from ', url_geo)
gdf = gpd.read_file(url_geo)
print('time taken', time.time()-t1, 'second')

gdf.plot()
plt.savefig('figpath.png', dpi=800, bbox_inches='tight')
```

```python
gdf1 = pd.DataFrame(columns = ['MSOA11CD_1', 'MSOA11NM_1', 'LONG_1', 'LAT_1'])
gdf1['MSOA11CD_1'] = gdf['MSOA11CD']
gdf1['MSOA11NM_1'] = gdf['MSOA11NM']
gdf1['LONG_1'] = gdf['LONG']
gdf1['LAT_1'] = gdf['LAT']
gdf1
```

```python
gdf2 = pd.DataFrame(columns = ['MSOA11CD_2', 'MSOA11NM_2', 'LONG_2', 'LAT_2'])
gdf2['MSOA11CD_2'] = gdf['MSOA11CD']
gdf2['MSOA11NM_2'] = gdf['MSOA11NM']
gdf2['LONG_2'] = gdf['LONG']
gdf2['LAT_2'] = gdf['LAT']
gdf2
```

```python
dfmg = pd.merge(df1,gdf1,left_on='Area of residence',right_on='MSOA11CD_1',how='inner')
dfmg
```

```python
dfmg1 = pd.merge(dfmg,gdf2,left_on='Area of workplace',right_on='MSOA11CD_2',how='inner')
dfmg1
```

```python
dfmg2 = dfmg1[['On foot', 'MSOA11CD_1', 'MSOA11NM_1', 'LONG_1', 'LAT_1', 'MSOA11CD_2', 'MSOA11NM_2', 'LONG_2', 'LAT_2']]
# dfmg2
```

```python
# df1=df1[~df1['A'].isin([1])]
dfmg3 = dfmg2[~dfmg2['On foot'].isin([0])]
dfmg3
```

```python
dfmg3.to_csv('tmp.csv',index=False)
```
