```python
from sklearn.preprocessing import RobustScaler, MinMaxScaler
rs = RobustScaler(quantile_range=(10.0, 90.0))

to_cluster = dfmg[['code', 'Usually resident population', 'PTAL', 'High qualification percentage', 'Annual household income', 'Older worker']].set_index('code').copy()
# to_cluster = dfmg[['Area code', "Work mainly at or from home", "Underground, metro, light rail, tram", 'Train', 'Bus, minibus or coach', "Taxi",
#                   "Motorcycle, scooter or moped", "Driving a car or van", "Passenger in a car or van", "Bicycle", "On foot"]].set_index('Area code').copy()
to_cluster.head()

# normnise the value
normed = to_cluster.copy()
for c in to_cluster.columns.values:
    normed[c] = rs.fit_transform(to_cluster[c].values.reshape(-1,1))
normed
```

```python
import pysal as ps
import numpy as np
import pandas as pd
import seaborn as sns
import geopandas as gpd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# calculate SSE for a range of number of cluster
list_SSE = []
min_k = 1
max_k = 10
range_k = range(min_k, max_k+1)
for i in range_k:
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(to_cluster)
    # inertia is a concept in physics. Roughly it means SSE of clustering.
    list_SSE.append(km.inertia_)

# plot
plt.plot(range_k, list_SSE, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.savefig('figpath.png', dpi=400, bbox_inches='tight')
plt.show()
# plt.savefig('figpath.png', dpi=800, bbox_inches='tight')
```

```python
# re-run the k-means process
from sklearn.cluster import KMeans
from sklearn import metrics
k_cluster = 4
random_seed = 1
kmeans_method = KMeans(n_clusters=k_cluster,random_state=random_seed)
kmeans_method.fit(normed)

gdf['Kmeans_Cluster'] = kmeans_method.labels_
fig_kmeans_4 = gdf.plot(column='Kmeans_Cluster', categorical=True, legend=True, figsize=(12,8),cmap = 'plasma')
from sklearn import metrics
print("Silhouette Score is :")
print(metrics.silhouette_score(normed, gdf['Kmeans_Cluster']))
plt.savefig('figpath.png', dpi=400, bbox_inches='tight')
```

```python
kmeans_label = pd.DataFrame(kmeans_method.labels_, columns = ['kmeans_label'])
kmeans_label
# to_cluster_labeled = pd.concat([to_cluster, DB_label], axis = 1).dropna()

# temp = to_cluster.join(kmeans_label)
to_cluster_labeled = to_cluster.copy()
to_cluster_labeled["kmeans_label"] = kmeans_method.labels_
to_cluster_labeled

summary_df_kmeans = to_cluster_labeled[['Usually resident population', 'PTAL', 'High qualification percentage', 'Annual household income', 'Older worker','kmeans_label']].groupby('kmeans_label').mean().round()

summary_df_kmeans['count'] = to_cluster_labeled.kmeans_label.value_counts()
summary_df_kmeans
```
